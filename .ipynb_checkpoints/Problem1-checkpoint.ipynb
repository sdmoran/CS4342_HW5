# Problem 1: Principal Component Analysis

### Problem 1A: Principal Components

```python
from scipy.io import loadmat
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# This is the data part of the .mat file
f = loadmat('data_hw5_1.mat')['R']

# PCA
pca = PCA(n_components=3)
principalComponents = pca.fit(f)
print("Principal components of X1, X2 and X3: ", pca.singular_values_)
```

### Problem 1B: Percentages of Variance

```python
print("Explained variance rations for X1, X2 and X3 (percentages): ", pca.explained_variance_ratio_)
```

### Problem 1C: Scatter Plots & Discussion
As is clearly visible in the scatterplots below, plots X2 vs X3 and X1 vs X3 display very little variance in any particular direction. However, in the plot of X1 vs X2, there is a very clear correlation between these two features. For this reason, X1 and X2 have the two highest contributions to the explained variance of the data overall. However, between these two variables, X1 has more variance, and therefore the vast majority of the varaince in the data is explained by X1.

```python
plt.figure()
plt.title("X1 vs X2")
plt.xlabel("X1")
plt.ylabel("X2")
plt.scatter(f[:, 0], f[:, 1])

plt.figure()
plt.title("X2 vs X3")
plt.xlabel("X2")
plt.ylabel("X3")
plt.scatter(f[:, 1], f[:, 2])

plt.figure()
plt.title("X1 vs X3")
plt.xlabel("X1")
plt.ylabel("X3")
plt.scatter(f[:, 0], f[:, 2])
```
